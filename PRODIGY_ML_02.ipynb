# Create a K-means clustering algorithm to group customers of a retail store based on their purchase history.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
data.shape
```

```python
data.info
```

```python
df1=data.iloc[:,2:5]
df1
```

```python
distortation=[]
k=range(1,20)
k
```

```python
from sklearn.cluster import KMeans
```

```python
for i in k:
    KMeanmodel=KMeans(n_clusters=i,max_iter=25)
    KMeanmodel.fit(df1)
    distortation.append(KMeanmodel.inertia_)
```

```python
plt.plot(figure=(16,8))
plt.plot(k,distortation,'bo-')
plt.xlabel('no of cluster')
plt.ylabel("WSS")
plt.title("Elbow method")
```

```python
df1.info()
```

```python
final_model1=KMeans(n_clusters=6,max_iter=25)
final_model1.fit(df1)
final_model1.cluster_centers_
final_model1.feature_names_in_
final_model1.labels_
```

```python
plt.scatter(df1['Spending Score (1-100)'],df1['Annual Income (k$)'],c=final_model1.labels_)
```

```python
final_model2=KMeans(n_clusters=7,max_iter=25)
final_model2.fit(df1)
final_model2.cluster_centers_
final_model2.feature_names_in_
final_model2.labels_
```

```python
plt.scatter(df1['Spending Score (1-100)'],df1['Annual Income (k$)'],c=final_model2.labels_)
```
